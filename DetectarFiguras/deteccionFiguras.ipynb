```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(title = "", image = None):

    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))

    plt.title(title)
    
    plt.show
```

## CIRCULO

```python
# Leer la imagen y convertirla a RGB
img = cv2.imread("img/circulo.jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Crear copias de la imagen para aislar cada canal
img_R = img_rgb.copy()
img_G = img_rgb.copy()
img_B = img_rgb.copy()

# Aislar el canal R
img_R[:, :, 1] = 0  # Green
img_R[:, :, 2] = 0  # Blue :(dos puntos es todas :(filas), :(columnas), R || G || B (QUITAR CANAL))

# Aislar el canal G
img_G[:, :, 0] = 0  # R
img_G[:, :, 2] = 0  # B

# Aislar el canal B
img_B[:, :, 0] = 0  # R
img_B[:, :, 1] = 0  # G

# Visualizar las imgenes
fig, ax = plt.subplots(1, 4, figsize=(20, 5))

ax[0].imshow(img_rgb)
ax[0].set_title("Imagen original")
ax[0].axis('off')

ax[1].imshow(img_R)
ax[1].set_title("Canal Rojo")
ax[1].axis('off')

ax[2].imshow(img_G)
ax[2].set_title("Canal Verde")
ax[2].axis('off')

ax[3].imshow(img_B)
ax[3].set_title("Canal Azul")
ax[3].axis('off')

plt.show()
```

```python
# Leer la imagen
img = cv2.imread("img/circulo.jpeg", cv2.IMREAD_GRAYSCALE)  # Leer en escala de grises directamente
green2gray = cv2.cvtColor(img_G, cv2.COLOR_RGB2GRAY)
# Calcular el histograma
hist = cv2.calcHist([green2gray], [0], None, [256], [0, 256])

# Visualizar la imagen y su histograma
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Mostrar la imagen en escala de grises
ax[0].imshow(green2gray, cmap='gray')
ax[0].set_title("Imagen en Grises")
ax[0].axis('off')

# Mostrar el histograma
ax[1].plot(hist, color='black')
ax[1].set_title("Histograma")
ax[1].set_xlabel("Intensidad de Pixel")
ax[1].set_ylabel("Cantidad de Pxeles")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Leer la imagen en escala de grises
green2gray = cv2.cvtColor(img_G, cv2.COLOR_RGB2GRAY)

# Definir el umbral
umbral = 29  # Puedes cambiar este valor segn tus necesidades

# Binarizar la imagen
_, img_binarizada = cv2.threshold(green2gray, umbral, 255, cv2.THRESH_BINARY)

# Visualizar la imagen original y la imagen binarizada
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Mostrar la imagen en escala de grises
ax[0].imshow(green2gray, cmap='gray')
ax[0].set_title("Imagen Original")
ax[0].axis('off')

# Mostrar la imagen binarizada
ax[1].imshow(img_binarizada, cmap='gray')
ax[1].set_title(f"Imagen Binarizada (Umbral = {umbral})")
ax[1].axis('off')

plt.tight_layout()
plt.show()
```
